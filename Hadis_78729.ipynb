```python
import requests
import json
from pprint import pprint
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import time
```

```python
#Search for all property types in zip code 78729 (Travis County-near Lake Travis)
payload = {
    'Accept': 'application/json',
    'apikey': "5a02104b4076981f3a2fbab6c6908160",
    'accept': 'application/json',
    'pagesize':'10'
}
market_value = []
sq_ft = []
zip_code = []
prop_type = []
year_built = []
for x in range(100):
    time.sleep(20)
    url = f"https://api.gateway.attomdata.com/propertyapi/v1.0.0/assessment/detail?postalcode=78729&pagesize=10&page={x}"
    s = requests.Session()
    response = s.get(url, headers = payload)
    print(response.url)
    content = response.content
    print(content)
    json_response = response.json()
    #print(response.content)
    try:
        for item in json_response['property']:
            market_value.append(item['assessment']['market']['mktttlvalue'])
            sq_ft.append(item['building']['size']['bldgsize'])
            zip_code.append(item['address']['postal1'])
            prop_type.append(item['summary']['proptype'])
            year_built.append(item['summary']['yearbuilt'])
    except:
        print("didnt work")
df_78729 = pd.DataFrame({"Zip Code": zip_code, "Property Type": prop_type, "Avg. Sq Ft": sq_ft, "Avg. Market Value": market_value,
                  "Avg. Year Built": year_built})
df_78729.to_csv('df_78729')
```
