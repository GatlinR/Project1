```python
import requests
import json
from pprint import pprint
import pandas as pd
from config import api_key
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Search for all property types in zip code 78745 (Travis County-near Lake Travis)
payload = {
    'Accept': 'application/json',
    'apikey': api_key,
    'accept': 'application/json'
}
url = "https://api.gateway.attomdata.com/propertyapi/v1.0.0/assessment/detail?postalcode=78745&pagesize=100"
s = requests.Session()
response = s.get(url, headers = payload)
json_response = response.json()
#pprint(json_response)
market_value = []
sq_ft = []
zip_code = []
prop_type = []
year_built = []
#print(response.content)
for item in json_response['property']:
    market_value.append(item['assessment']['market']['mktttlvalue'])
    sq_ft.append(item['building']['size']['bldgsize'])
    zip_code.append(item['address']['postal1'])
    prop_type.append(item['summary']['proptype'])
    year_built.append(item['summary']['yearbuilt'])
df_78745 = pd.DataFrame({"Zip Code": zip_code, "Property Type": prop_type, "Avg. Sq Ft": sq_ft, "Avg. Market Value": market_value,
                  "Avg. Year Built": year_built})
#Add Price per SqFt column 
df_78745["Avg. Price per SqFt"] = round(df_78745["Avg. Market Value"]/df_78745["Avg. Sq Ft"],2)
#Get rid of rows with zero in data
df_78745 = df_78745.loc[df_78745["Avg. Sq Ft"]!=0,:]
df_78745 = df_78745.loc[df_78745["Avg. Year Built"]!=0,:]
df_78745 = df_78745.loc[df_78745["Avg. Market Value"]!=0,:]
#Narrow down dataframe to propery types we want to look at
df_78745 = df_78745.loc[(df_78745["Property Type"]=="CONDOMINIUM")|(df_78745["Property Type"]=="DUPLEX")|
                        (df_78745["Property Type"]=="SFR"),:]
df_78745

#Calculate average price per sqft for Condominium, Duplex, SFR, and Family Dwelling-Ariana
property_df = df_78745.groupby("Property Type")
property_df.mean()
#Find average house value (SFR), plot Year Built vs. Market Value of SFR on line graph to see how it changes over time
#Does age of house affect its market value? -Gatlin

#Find count of each property type, what property types are most popular?

#Find count of Condominium, Duplex, SFR, and Family Dwelling based on year built-Angie
```
